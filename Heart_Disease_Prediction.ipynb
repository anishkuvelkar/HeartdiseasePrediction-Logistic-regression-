#Data preprocessing

```python
DS.head()
```

```python
DS.groupby('Heart Disease').mean()
```

```python
DS['Heart Disease'].value_counts()
```

```python
lastcolums = DS['Heart Disease']
```

```python
lastcolums = lastcolums.replace({'Presence': 1, 'Absence': 0})
```

```python
DS['Target'] = lastcolums
```

```python
DS.head()
```

```python
DS = DS.drop(columns='Heart Disease', axis=1)
DS.head()
```

```python
DS.isnull().sum()
```

1-> Has a heart disease
2-> Does not have a heart disease

#Splitting data into training and test

```python
X = DS.drop(columns='Target', axis=1)
Y = DS['Target']
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
Y.shape, Y_train.shape, Y_test.shape
```

#Model

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_Predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_Predict, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on Test data : ', test_data_accuracy)
```

#Building a prediction model

```python
input = np.array([67,0,3,115,564,0,2,160,0,1.6,2,0,7])
input = input.reshape(1,-1)
prediction = model.predict(input)
print(prediction)
if (prediction[0]== 0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```
